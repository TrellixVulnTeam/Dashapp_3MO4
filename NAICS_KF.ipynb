Below is where I found the urls to loop through. I did this by searching for each integer (1-9; 0 wasn't available) and then finding the max number of results for that search. Each url is displayed 40 records at a time, so I updated each url by adding 40 and then saving the new url to a list


```
base1_url  = 'https://www.sec.gov/cgi-bin/browse-edgar?action=getcompany&company=1&owner=exclude&match=&start='
i = 0
one_urls= []
while i < 1720:
  i = i+40
  one_urls.append(base1_url+str(i))
```
